```python
import tkinter as tk
from tkinter import ttk, messagebox
from tkcalendar import DateEntry
import pandas as pd
import gspread
from gspread_dataframe import set_with_dataframe
from datetime import datetime

# üîπ ƒê·ªçc danh s√°ch m√£ h√†ng t·ª´ file Excel
file_excel = r"C:\Users\admin\Downloads\MA HANG SKU.xlsx"
try:
    df_excel = pd.read_excel(file_excel, engine='openpyxl')
    danh_sach_ma_hang = df_excel["M√£ h√†ng"].astype(str).str.strip().dropna().tolist()
except Exception as e:
    messagebox.showerror("L·ªói", f"Kh√¥ng th·ªÉ ƒë·ªçc file Excel: {e}")
    danh_sach_ma_hang = []

# üîπ K·∫øt n·ªëi Google Sheets
gc = gspread.service_account(filename=r"C:\Users\admin\Downloads\api-map-ha-b94f9c05f9f2.json")
sheet_url = "https://docs.google.com/spreadsheets/d/17bY2a1xeuvuJbgy4-hqMtxLPGcVSXr5haAMS9heqGmQ/edit#gid=0"
sh = gc.open_by_url(sheet_url)
worksheet = sh.sheet1  # Ch·ªçn sheet ƒë·∫ßu ti√™n

# üîπ T·∫°o c·ª≠a s·ªï ch√≠nh
root = tk.Tk()
root.title("Nh·∫≠p li·ªáu m√£ h√†ng")
root.geometry("1000x500")
root.configure(bg="#d4edda")  # M√†u n·ªÅn xanh l√° nh·∫°t

# üîπ Ti√™u ƒë·ªÅ l·ªõn
title_label = tk.Label(root, text="NH·∫¨P XU·∫§T X√Å ƒê√ìNG G√ìI - XU·∫§T TP T·ª™ ƒê√ìNG G√ìI", font=("Arial", 16, "bold"), bg="#155724", fg="white")
title_label.pack(pady=10)

# üîπ T·∫°o √¥ nh·∫≠p li·ªáu
frame = tk.Frame(root, bg="#d4edda")
frame.pack(pady=10)

labels = ["Ng√†y:", "M√£ h√†ng:", "Xu·∫•t ƒë√≥ng g√≥i:", "Xu·∫•t th√†nh ph·∫©m:", "Xu·∫•t h·ªßy:", "Ghi ch√∫:"]
for i, label in enumerate(labels):
    tk.Label(frame, text=label, bg="#d4edda", fg="#155724", font=("Arial", 10, "bold")).grid(row=0, column=i, padx=5)

# üîπ Bi·∫øn nh·∫≠p li·ªáu
ngay_var = tk.StringVar()
ma_hang_var = tk.StringVar()
xuat_dong_goi_var = tk.IntVar()
xuat_thanh_pham_var = tk.IntVar()
xuat_huy_var = tk.IntVar()
ghi_chu_var = tk.StringVar()

# üîπ Tr∆∞·ªùng nh·∫≠p ng√†y
ngay_entry = DateEntry(frame, textvariable=ngay_var, date_pattern="yyyy-MM-dd")
ngay_entry.grid(row=1, column=0, padx=5)
ngay_entry.set_date(datetime.today())

# üîπ Tr∆∞·ªùng nh·∫≠p m√£ h√†ng (r·ªông g·∫•p ƒë√¥i)
ma_hang_dropdown = ttk.Combobox(frame, textvariable=ma_hang_var, values=danh_sach_ma_hang, width=30)
ma_hang_dropdown.grid(row=1, column=1, padx=5)

xuat_dong_goi_entry = tk.Entry(frame, textvariable=xuat_dong_goi_var, width=10)
xuat_dong_goi_entry.grid(row=1, column=2, padx=5)

xuat_thanh_pham_entry = tk.Entry(frame, textvariable=xuat_thanh_pham_var, width=10)
xuat_thanh_pham_entry.grid(row=1, column=3, padx=5)

xuat_huy_entry = tk.Entry(frame, textvariable=xuat_huy_var, width=10)
xuat_huy_entry.grid(row=1, column=4, padx=5)

ghi_chu_entry = tk.Entry(frame, textvariable=ghi_chu_var, width=20)
ghi_chu_entry.grid(row=1, column=5, padx=5)

# üîπ H√†m c·∫≠p nh·∫≠t d·ªØ li·ªáu l√™n Google Sheets
def update_google_sheets():
    ngay = ngay_var.get()
    ma_hang = ma_hang_var.get()
    xuat_dong_goi = xuat_dong_goi_var.get()
    xuat_thanh_pham = xuat_thanh_pham_var.get()
    xuat_huy = xuat_huy_var.get()
    ghi_chu = ghi_chu_var.get()

    if not ma_hang:
        messagebox.showerror("L·ªói", "Vui l√≤ng ch·ªçn m√£ h√†ng!")
        return

    df = pd.DataFrame([[ngay, ma_hang, xuat_dong_goi, xuat_thanh_pham, xuat_huy, ghi_chu]],
                      columns=["Ng√†y", "M√£ h√†ng", "Xu·∫•t ƒë√≥ng g√≥i", "Xu·∫•t th√†nh ph·∫©m", "Xu·∫•t h·ªßy", "Ghi ch√∫"])
    worksheet.append_row(df.values.tolist()[0])
    messagebox.showinfo("Th√¥ng b√°o", "D·ªØ li·ªáu ƒë√£ ƒë∆∞·ª£c c·∫≠p nh·∫≠t l√™n Google Sheets!")

# üîπ T·∫°o n√∫t c·∫≠p nh·∫≠t Google Sheets
btn_frame = tk.Frame(root, bg="#d4edda")
btn_frame.pack(pady=10)
update_btn = tk.Button(btn_frame, text="C·∫≠p nh·∫≠t Google Sheets", command=update_google_sheets, bg="#28a745", fg="white", font=("Arial", 10, "bold"))
update_btn.grid(row=0, column=0, padx=5)

# üî• Ch·∫°y ·ª©ng d·ª•ng
root.mainloop()
```
